```python
df = df.drop(columns=['wage_euro', 'preferred_foot', 'international_reputation(1-5)', 'weak_foot(1-5)', 'skill_moves(1-5)'])  # Remove a coluna pelo nome
```

```python
df = df.drop(columns=['national_jersey_number', 'curve', 'crossing', 'interceptions', 'vision', 'marking', 'standing_tackle', 'sliding_tackle'])  # Remove a coluna pelo nome
```

```python
df.info()
```

```python
df = df.drop(columns=['jumping', 'long_passing', 'freekick_accuracy', 'national_team_position', 'national_rating', 'national_team', 'release_clause_euro', 'body_type'])  # Remove a coluna pelo nome
```

```python
df.info()
```

```python
df['velocidade'] = (df['acceleration'] + df['sprint_speed']) / 2
```

```python
df.info()
```

```python
df = df.drop(columns=['acceleration', 'sprint_speed'])  # Remove a coluna pelo nome
```

```python
df['finalizacao'] = (df['penalties'] + df['long_shots'] + df['shot_power'] + df['volleys'] + df['heading_accuracy'] + df['finishing'] ) / 6
```

```python
df = df.drop(columns=['penalties', 'long_shots', 'shot_power', 'volleys', 'heading_accuracy', 'finishing'])  # Remove a coluna pelo nome
```

```python
df.info()
```

```python
df['drible'] = (df['dribbling'] + df['ball_control'] + df['agility'] + df['reactions'] + df['balance']) / 5
```

```python
df = df.drop(columns=['dribbling', 'ball_control', 'agility', 'reactions', 'balance'])  # Remove a coluna pelo nome
```

```python
df.info()
```

```python
df['fisico'] = (df['stamina'] + df['strength']) / 2
```

```python
print(df.info())
```

```python
df = df.drop(columns=['name', 'age','stamina','strength'])  # Remove a coluna pelo nome
```
